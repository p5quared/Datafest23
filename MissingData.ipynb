```python
import pandas as pd

clients = pd.read_csv('data/clients.csv')
'''
3 types of missing data:
NA, NULL, I'd prefer not to answer.

Categorical data -> "I'd prefer not to answer"
Numeric data -> NULL
'''
clients.loc[:,'EthnicIdentity'].fillna('I\'d rather not answer', inplace=True)

clients
```

```python
questions = pd.read_csv('data/questions.csv')
client_questions = clients.merge(questions, how='outer', left_on='ClientUno', right_on='AskedByClientUno')
# df.drop, suffix

client_questions
```

```python
client_questions.value_counts('EthnicIdentity')[:5].plot(kind='bar')
```
